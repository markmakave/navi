```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../../build/interpolation.csv')
```

```python
plt.figure(figsize=(10,5))

plt.scatter(df['x'], df['y'], s=1, c=df['x'].map(lambda x: 'blue' if (x < 0 or x > 1) else 'red'))

plt.axvline(x=0, c='red')
plt.axvline(x=0.375, c='red')
plt.axvline(x=0.625, c='red')
plt.axvline(x=1, c='red')

plt.axvline(x=-0.125, c='lime')
plt.axvline(x=0.125, c='lime')
plt.axvline(x=0.875, c='lime')
plt.axvline(x=1.125, c='lime')
```
